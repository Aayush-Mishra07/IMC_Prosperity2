```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
pricesday_1=pd.read_csv('/Users/aayushmishra/Desktop/IMCProsperity/Data/round-2-island-data-bottle/prices_round_2_day_-1.csv',delimiter=';')
pricesday0=pd.read_csv('/Users/aayushmishra/Desktop/IMCProsperity/Data/round-2-island-data-bottle/prices_round_2_day_0.csv',delimiter=';')
pricesday1=pd.read_csv('/Users/aayushmishra/Desktop/IMCProsperity/Data/round-2-island-data-bottle/prices_round_2_day_1.csv',delimiter=';')
df = pd.concat([pricesday_1,pricesday0,pricesday1], ignore_index=True)

df["HUMIDITY"]
```

```python
sunlight = df['SUNLIGHT'].to_numpy()
humidity = df['HUMIDITY'].to_numpy()
orchids = df['ORCHIDS'].to_numpy()
```

```python
scaler = MinMaxScaler()
sun = scaler.fit_transform(sunlight.reshape(-1, 1))
hum = scaler.fit_transform(humidity.reshape(-1, 1))
orc = scaler.fit_transform(orchids.reshape(-1, 1))

variable = (0.5*(sun**0.72) + 16277/30000*(hum**0.71))
var = scaler.fit_transform(variable.reshape(-1, 1))
```

```python
plt.figure(figsize=(15, 5))
# plt.plot(sun[0:10001], label="sun")
# plt.plot(hum[0:10001], label='hum')
plt.plot(orc[0:10001], label='orc')
plt.plot(var[0:10001], label='var')

plt.legend()
plt.show()
```

```python
# Concatenate arrays along the second axis (columns)
combined_array = np.vstack((var.reshape(-1), orc.reshape(-1)))

# Compute correlation matrix
correlation_matrix = np.corrcoef(combined_array)

print("Correlation Matrix:")
print(correlation_matrix)
```
